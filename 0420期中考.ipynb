```
install.packages("ggplot2")
```

```
library(ggplot2)
```

```
NBA1617 <- data.frame(gp = factor(rep(letters[1:3],each = 10)),y = rnorm(30))
ggplot(NBA1617,aes(x = gp, y = y)) + geom_point()
```

```
ggplot(NBA1617,aes(x =gp,y = y)) +geom_boxplot()
```

```
NBA1617$z<-NBA1617$y+rnorm(30)
ggplot(NBA1617,aes(x=z,y=y))+geom_point()+facet_grid(gp~.)
ggplot(NBA1617,aes(x=z,y=y))+geom_point()+facet_grid(.~gp)
```

```
ggplot(NBA1617,aes(x=z,y=y))+geom_point()+facet_grid(gp~.)+geom_smooth()
ggplot(NBA1617,aes(x=z,y=y))+geom_point()+facet_grid(.~gp)+geom_smooth()
```

```
ggplot(NBA1617,aes(x=z,y=y))+geom_line()+facet_grid(gp~.)
ggplot(NBA1617,aes(x=z,y=y,color=gp))+geom_line()
```
